```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import os

cwd = os.getcwd()
root_path = os.path.abspath(os.path.join(cwd, os.pardir))
sys.path.append(root_path)
```

```python
from implementation.eigen_value_decomposition import eigen_decomposition_full, eigen_decomposition_real_full

# Define a 3x3 symmetric matrix
A = np.array([[4, 2, 1],
              [2, 5, 3],
              [1, 3, 6]])

# Perform eigen decomposition for symmetric matrix
P, D, P_inv = eigen_decomposition_real_full(A)

print("Matrix A:")
print(A)
print("\nEigenvectors (P):")
print(P)
print("\nEigenvalue matrix (D):")
print(D)

# Verify A = P @ D @ P_inv
A_reconstructed = P @ D @ P_inv
print("\nReconstructed A = P @ D @ P_inv:")
print(A_reconstructed)
print("\nReconstruction correct:", np.allclose(A, A_reconstructed))
```

```python
eigenvalues = np.diag(D)

plt.figure(figsize=(8, 5))
plt.bar(range(len(eigenvalues)), eigenvalues, color=['steelblue', 'coral', 'seagreen'])
plt.title('Eigenvalues of Symmetric Matrix A')
plt.xlabel('Index')
plt.ylabel('Eigenvalue')
plt.xticks(range(len(eigenvalues)), [f'Î»{i+1}' for i in range(len(eigenvalues))])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
