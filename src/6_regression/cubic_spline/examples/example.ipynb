## Notes Worked Example

Points: (0, 0), (1, 0.5), (2, 0)  
Natural BCs: M₀ = M₂ = 0, solving gives M₁ = −1.5  
S₀(0.5) = 0.34375

```python
from implementation.cubic_spline import cubic_spline

x = np.array([0.0, 1.0, 2.0])
y = np.array([0.0, 0.5, 0.0])

spline = cubic_spline(x, y)
ref = CubicSpline(x, y, bc_type='natural')

print('Notes Worked Example: (0,0), (1,0.5), (2,0)')
print(f'{"Point":>8}  {"Impl":>10}  {"scipy":>10}')
for pt in [0.0, 0.25, 0.5, 0.75, 1.0, 1.25, 1.5, 1.75, 2.0]:
    impl_val = spline(pt)
    scipy_val = float(ref(pt))
    print(f'{pt:8.2f}  {impl_val:10.6f}  {scipy_val:10.6f}')
print(f'\nS(0.5) = {spline(0.5):.5f}  (notes: 0.34375)')
```

## sin(x) — Implementation vs scipy Surface

```python
x_data = np.linspace(0, 2*np.pi, 8)
y_data = np.sin(x_data)

spline = cubic_spline(x_data, y_data)
ref = CubicSpline(x_data, y_data, bc_type='natural')

x_fine = np.linspace(x_data[0], x_data[-1], 300)
y_impl = np.array([spline(xi) for xi in x_fine])
y_scipy = ref(x_fine)
y_exact = np.sin(x_fine)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 7), height_ratios=[3, 1])

ax1.scatter(x_data, y_data, s=80, zorder=5, color='black', label='Data')
ax1.plot(x_fine, y_impl, lw=2, label='Implementation')
ax1.plot(x_fine, y_scipy, '--', lw=2, label='scipy (natural)')
ax1.plot(x_fine, y_exact, ':', lw=1, alpha=0.5, label='sin(x) exact')
ax1.set_ylabel('y'); ax1.set_title('Cubic Spline: sin(x)')
ax1.legend(); ax1.grid(True, alpha=0.3)

ax2.plot(x_fine, np.abs(y_impl - y_scipy), color='red')
ax2.set_xlabel('x'); ax2.set_ylabel('|impl - scipy|')
ax2.set_title(f'Max difference = {np.abs(y_impl - y_scipy).max():.2e}')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
