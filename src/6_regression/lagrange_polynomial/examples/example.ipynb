## Notes Worked Example

Points: A(−1, 1), B(2, 3), C(3, 5)  →  L(x) = x²/3 + x/3 + 1

```python
from implementation.lagrange_polynomial import lagrange_polynomial

x = np.array([-1.0, 2.0, 3.0])
y = np.array([1.0, 3.0, 5.0])

print('Notes Worked Example: A(-1,1), B(2,3), C(3,5)')
print(f'{"Point":>10}  {"Impl":>10}  {"scipy":>10}  {"Exact":>10}')
poly = scipy_lagrange(x, y)
for pt in [-1, 0, 1, 2, 3]:
    impl_val = lagrange_polynomial(x, y, float(pt))
    scipy_val = float(poly(pt))
    exact = pt**2/3 + pt/3 + 1
    print(f'{pt:10.1f}  {impl_val:10.6f}  {scipy_val:10.6f}  {exact:10.6f}')
```

## Larger Example — Implementation vs scipy

5 scattered points with curve overlay and difference plot.

```python
x = np.array([-9, -5, -2.5, 4, 7], dtype=float)
y = np.array([-2, 3, 0, 5, 11], dtype=float)

x_domain = np.linspace(1.1 * x.min(), 1.1 * x.max(), 200)
y_impl = np.array([lagrange_polynomial(x, y, xi) for xi in x_domain])
y_scipy = scipy_lagrange(x, y)(x_domain)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 7), height_ratios=[3, 1])

ax1.scatter(x, y, s=120, zorder=5, color='black', label='Data')
ax1.plot(x_domain, y_impl, lw=2, label='Implementation')
ax1.plot(x_domain, y_scipy, '--', lw=2, label='scipy.interpolate.lagrange')
ax1.set_ylabel('y')
ax1.set_title('Lagrange Polynomial Interpolation')
ax1.legend()
ax1.grid(True, alpha=0.3)

ax2.plot(x_domain, np.abs(y_impl - y_scipy), color='red')
ax2.set_xlabel('x')
ax2.set_ylabel('|difference|')
ax2.set_title(f'Max |impl - scipy| = {np.abs(y_impl - y_scipy).max():.2e}')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
