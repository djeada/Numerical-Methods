## Gaussian Central-Difference Interpolation

This notebook demonstrates Gauss's forward and backward central-difference
interpolation on equally-spaced data and verifies the results against
`scipy.interpolate.BarycentricInterpolator` (which computes the unique
polynomial interpolant).

```python
from implementation.gaussian_interpolation import gaussian_interpolation
from scipy.interpolate import BarycentricInterpolator

# Data from the notes example
x_data = np.array([0.0, 1.0, 2.0, 3.0, 4.0])
y_data = np.array([2.0, 3.5, 5.0, 5.8, 6.0])

test_points = np.array([0.5, 1.5, 2.5, 3.5])
interpolated = np.array([gaussian_interpolation(x_data, y_data, p) for p in test_points])
scipy_ref = BarycentricInterpolator(x_data, y_data)(test_points)

print('Gaussian Central-Difference Interpolation Results:')
print(f'{"x":>6}  {"Gauss":>12}  {"scipy":>12}  {"match":>6}')
for p, g, s in zip(test_points, interpolated, scipy_ref):
    print(f'{p:6.1f}  {g:12.6f}  {s:12.6f}  {np.isclose(g, s):>6}')
```

```python
# sin(x) example
x_sin = np.array([0.0, 1.0, 2.0, 3.0, 4.0])
y_sin = np.sin(x_sin)

test_pts_sin = np.array([0.5, 1.5, 2.5, 3.5])
interp_sin = np.array([gaussian_interpolation(x_sin, y_sin, p) for p in test_pts_sin])

print('\nsin(x) Interpolation:')
print(f'{"x":>6}  {"Gauss":>12}  {"sin(x)":>12}  {"error":>12}')
for p, g in zip(test_pts_sin, interp_sin):
    print(f'{p:6.1f}  {g:12.6f}  {np.sin(p):12.6f}  {abs(g - np.sin(p)):12.2e}')
```

```python
x_fine = np.linspace(0, 4, 200)
y_fine = np.array([gaussian_interpolation(x_data, y_data, p) for p in x_fine])
y_scipy = BarycentricInterpolator(x_data, y_data)(x_fine)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: notes example data
ax = axes[0]
ax.plot(x_fine, y_fine, 'b-', label='Gauss interpolation')
ax.plot(x_fine, y_scipy, 'r--', alpha=0.7, label='scipy (verification)')
ax.scatter(x_data, y_data, color='red', s=100, zorder=5, label='Data points')
ax.scatter(test_points, interpolated, color='green', s=80, zorder=5, marker='x', label='Evaluated points')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Notes example data')
ax.legend()
ax.grid(True)

# Right: sin(x)
x_fine_sin = np.linspace(0, 4, 200)
y_fine_sin = np.array([gaussian_interpolation(x_sin, y_sin, p) for p in x_fine_sin])
ax = axes[1]
ax.plot(x_fine_sin, np.sin(x_fine_sin), 'k--', alpha=0.5, label='sin(x) exact')
ax.plot(x_fine_sin, y_fine_sin, 'b-', label='Gauss interpolation')
ax.scatter(x_sin, y_sin, color='red', s=100, zorder=5, label='Data points')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('sin(x) interpolation')
ax.legend()
ax.grid(True)

plt.tight_layout()
plt.show()
```
