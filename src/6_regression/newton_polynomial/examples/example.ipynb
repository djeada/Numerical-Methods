```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import os

cwd = os.getcwd()
root_path = os.path.abspath(os.path.join(cwd, os.pardir))
sys.path.append(root_path)
```

```python
from implementation.newton_polynomial import newton_polynomial

x_data = np.linspace(0, np.pi, 5)
y_data = np.sin(x_data)

test_points = [0.5, 1.0, 1.5, 2.0, 2.5]
print("Newton Polynomial Interpolation of sin(x):")
for p in test_points:
    val = newton_polynomial(x_data, y_data, p)
    print(f"  x = {p:.1f}: interpolated = {val:.6f}, sin(x) = {np.sin(p):.6f}")
```

```python
x_fine = np.linspace(0, np.pi, 200)
y_fine = np.array([newton_polynomial(x_data, y_data, p) for p in x_fine])

plt.figure(figsize=(10, 6))
plt.plot(x_fine, np.sin(x_fine), 'k--', label='sin(x)', alpha=0.5)
plt.plot(x_fine, y_fine, 'b-', label='Newton Polynomial')
plt.scatter(x_data, y_data, color='red', s=100, zorder=5, label='Data Points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Newton Polynomial Interpolation of sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```
