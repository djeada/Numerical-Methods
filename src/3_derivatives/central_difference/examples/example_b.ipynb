```python
import numpy as np
import matplotlib.pyplot as plt

# define the function
def f(x):
    return x ** 3


# define the domain and plot the function
x = np.linspace(-0.2, 0.8, 1000)
plt.plot(x, f(x))

# choose point at which we want to calculate the derivative and the step
x0 = 0.3
x1 = x0 + 0.5
xr = x0 + 0.3
xl = x0 - 0.3

# plot the central difference
plt.plot([xl, xr], [f(xl), f(xr)], "r", label="Central difference")

# define the exact derivative and plot it
def df(x):
    return 3 * x ** 2


h = 0.2
plt.plot(
    [x0 - h, x0 + h], [f(x0) - h * df(x0), f(x0) + h * df(x0)], label="Exact derivative"
)

# add axes and labels
plt.xticks((xl, x0, xr))
plt.plot([xl, xl], [-0.1, f(xl)], "k:")
plt.plot([xr, xr], [-0.1, f(xr)], "k:")
plt.plot([x0, x0], [-0.1, f(x0)], "k:")

plt.yticks((f(xl), f(xr)))
plt.plot([-0.3, xl], [f(xl), f(xl)], "k:")
plt.plot([-0.3, xr], [f(xr), f(xr)], "k:")

# Add title and show the plots
plt.title("Central difference method", fontsize=16)
plt.xlabel("$x$", fontsize=16)
plt.ylabel("$f(x) = x^3$", fontsize=16)
plt.legend(loc=2)
plt.tight_layout()
plt.show()
```
