```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the function
def f(x):
    return x ** 3


# define the domain and plot the function
x = np.linspace(0, 1, 1000)
plt.plot(x, f(x))

# choose point at which we want to calculate the derivative and the step
x0 = 0.3
x1 = x0 + 0.5

# plot the forward difference
plt.plot([x0, x1], [f(x0), f(x1)], "r", label="Forward difference")

# define the exact derivative and plot it
def df(x):
    return 3 * x ** 2


h = 0.2
plt.plot(
    [x0 - h, x0 + h], [f(x0) - h * df(x0), f(x0) + h * df(x0)], label="Exact derivative"
)

# add axes and labels
plt.xticks((x0, x1), label=("$x_0$", "$x_0 + \Delta x$"))
plt.plot([x0, x0], [-0.1, f(x0)], "g:")
plt.plot([x1, x1], [-0.1, f(x1)], "g:")

plt.yticks((f(x0), f(x1)), label=("$f(x_0)$", "$f(x_0 + \Delta x)$"))
plt.plot([-0.1, x0], [f(x0), f(x0)], "g:")
plt.plot([-0.1, x1], [f(x1), f(x1)], "g:")

# Add title and show the plots
plt.title("Forward difference method", fontsize=16)
plt.xlabel("$x$", fontsize=16)
plt.ylabel("$f(x) = x^3$", fontsize=16)
plt.legend(loc=2)
plt.tight_layout()
plt.show()
```
