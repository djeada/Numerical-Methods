```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**3

x_min = 0
x_max = np.pi*2
x = np.linspace(x_min, x_max, 100)
y = f(x)
plt.plot(x, y, label=r'$f(x) = x^3$')

# midpoint rule
n = 10
step = (x_max - x_min) / n
for i in range(n):
    x_i = x_min + i * step
    x_i1 = x_min + (i + 1) * step
    x_mid = (x_i + x_i1) / 2
    y_mid = f(x_mid)
    plt.fill_between([x_i, x_i1], [y_mid, y_mid], color='red', alpha=0.2)

plt.title('Midpoint rule')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best')

plt.show()
```
